```python
data_0 = data[data['1024'] == 0 ]
data_0.to_csv('notA.csv' , index=False)

data_1 = data[data['1024'] != 0]
data_1.to_csv('A.csv' , index=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test = train_test_split(data_0, test_size=0.2)
x_train.to_csv('notA_train.csv' , index=False)
x_test.to_csv('notA_test.csv', index=False)
```

```python
y_train, y_test = train_test_split(data_1, test_size=0.2)
y_train.to_csv('A_train.csv' , index=False)
y_test.to_csv('A_test.csv', index=False)
```

```python
import neurokit2 as nk
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm

sampling_rate = 512
num_signals = x_train.shape[0]
batch_size = 64

for i in tqdm(range(0, num_signals, batch_size)):
    batch_signals = x_train.iloc[i:i+batch_size, :-1]
    
    for j, signal in enumerate(batch_signals.values):
        f, t, cwtm = nk.signal_timefrequency(signal, sampling_rate, max_frequency=32, method="pwvd", show=False)
        
        plt.imshow(cwtm, aspect='auto', cmap='jet', origin='lower')
        plt.axis('off')
        plt.savefig(f"./images2/notA_train/cwtm_image_{i + j}.png", bbox_inches='tight', pad_inches=0)
        plt.close()
```
